```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('data_fifa.csv')
print(df.columns)
```

```python
df
```

```python
sns.countplot(data = df, x = 'Real Face')
```

```python
sns.color_palette()
```

```python
color_base = sns.color_palette()[0]
```

```python
sns.countplot(data = df, x = 'Real Face', color = color_base)
```

```python
sns.countplot(data = df, x = 'Position')
plt.xticks(rotation = 90)
```

```python
order_position = df['Position'].value_counts().index

sns.countplot(data = df, x = 'Position', color=color_base, order=order_position)
plt.xticks(rotation = 90)
```

```python
print(df.columns)
```

```python
sorted_values = df['Real Face'].value_counts()
print(sorted_values)
print(type(sorted_values))
plt.pie(sorted_values, labels = sorted_values.index, startangle=90, counterclock=False,  autopct = '%1.1f%%')

plt.axis('square')
```

```python
sorted_values = df['Real Face'].value_counts()
print(sorted_values)
print(type(sorted_values))
plt.pie(sorted_values, labels = sorted_values.index, startangle=90, counterclock=False,  autopct = '%1.1f%%', wedgeprops={'width':0.5, 'edgecolor':'w'})

plt.axis('square')
```

## Histogramas

```python
import numpy as np

print(df['Potential'].describe())

bins_base = np.arange(45, df['Potential'].max() + 10, 3)
print(bins_base)
plt.hist( data = df, x = 'Potential', bins = bins_base);
```

```python
sns.distplot(df['Potential'], kde=False)
```

```python
plt.scatter(data = df, x = 'BallControl', y = 'Dribbling')
plt.xlabel('BallControl')
plt.ylabel('Dribbling')
```

```python
sns.regplot(data = df, x = 'BallControl', y = 'Dribbling')
plt.xlabel('BallControl')
plt.ylabel('Dribbling')
```

```python
bins_potential = np.arange(45, df['Potential'].max() + 5, 5)
bins_age = np.arange(15, df['Age'].max() + 5, 5)

plt.hist2d( data = df, x = 'Potential', y = 'Age', bins = [bins_potential, bins_age] )
plt.xlabel('Potential')
plt.ylabel('Age')
plt.colorbar();
```
